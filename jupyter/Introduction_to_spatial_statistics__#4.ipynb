#### Tipi di dati spaziali (geostatistical, lattice, point patterns)
La classificazione più grossolana che è possibile operare sui dati spaziali è quella proposta da Cressie (1993) in cui i tipi di dati sono distinti in funzione della natura del domninio spaziale.<br>
In termini matematici:<br>
$\{Z(\mathbf{s}) \, : \, \mathbf{s} \in\mathfrak{D}\subset\Re^d \}$
ove
<ul>
  <li>$d$ è la dimensionalità con cui si osserva la variabile</li>
  <li>$Z$ denota l'attributo o la variabile di osservazione/misura (es. produzione, contaminazione, C.O., ...)</li>
  <li>$\mathbf{s}$ è la posizione geografica/geospaziale in cui il variabile $Z$ è osservata.<br>
  $\mathbf{s}$ è definita dal vettore di coordinate $(d\times1)$</li>
  <ul>
    <li>$d=1$, transetto</li>
    <li>$d=2$, superficie Cartesiana del tipo $\mathbf{s}=[x,y]$</li>
    <li>$d=3$, volume, nel caso del suolo $\mathbf{s}=[x,y,z]$ ove $z=soil\,depth$</li>
  </ul><br>
  <li>$\mathfrak{D}$ è il dominio spaziale in cui la variabile di misura si distribuisce </li>
</ul>

#### Esempio di superficie Cartesiana del tipo $\mathbf{s}=[x,y]$:

```r
d = 2;
x = [7,5,1,9,3,11,6]';
y = [4,1,10,6,5,8,9]';
s = [x,y];

% s(x,y)
for ii = 1:numel(x), Si{ii,1} = ['S_',num2str(ii),'(',num2str(x(ii)),',',num2str(y(ii)),')']; end

% FIGURE:
figure
scatter(x,y,450,'.','r'),axis([0,13,0,12])
xlabel('Easting [m]','FontWeight','b','FontSize',12)
ylabel('Northing [m]','FontWeight','b','FontSize',12)
title('d=2, s=[x,y]','FontWeight','b','FontSize',14)
text(x,y,Si,'HorizontalAlignment','left','VerticalAlignment','bottom','FontWeight','b','FontSize',12)
box on
```
