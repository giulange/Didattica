### Inverso della distanza

```r
library(gstat) # carica il pacchetto GSTAT
```

# $z_u\left(p\right) = \frac{\sum _{i=1}^{n}\left(\frac{z_i}{d_i^p}\right)}{\sum _{i=1}^{n}\left(\frac{1}{d_i^p}\right)}$

```r
# interpolazione inverso della distanza
zinc.idw = idw(zinc~1,     # come interpolare Zn? (~1 : senza info ausiliarie)
               meuse,      # dov'è Zn? (in meuse)
               meuse.grid) # dove interpolare? (nei pixel meuse.grid)
class(zinc.idw) # stampa il tipo di dato restituito da GSTAT
```

```r
options(repr.plot.width=4)
spplot(zinc.idw["var1.pred"],          # mappa di Zn (pred)
       main = "zinc IDW interpolation")# titolo della mappa
```

### Variografia (ossia analisi geostatistica dei dati per mezzo del variogramma)

Comparando il grafico a bolle delle misure di <b>zinco</b> con la mappa della <b>distanza dal fiume</b>, diventa evidente che le concentrazioni maggiori del metallo sono misurate in prossimità del fiume.<br>

Seguono la correlazione e lo scatterplot con linea di tendenza sui dati grezzi:

```r
options(repr.plot.width=5,repr.plot.height=4)
cor_raw = cor(cbind(meuse$zinc,meuse$dist))
plot(zinc~dist, meuse)           # scatter dei punti Zn vs DIST
abline(lm(zinc ~ dist, meuse))   # linea di tendenza
title( paste("Correlazione ",round(cor_raw[2],3)) )
```

<b>Il grafico sopra evidenzia un andamento non lineare.</b><br>
Questa relazione può essere linearizzata mediante una <b>trasformazione</b> logaritmica delle concentrazioni di zinco, e prendendo la radice quadrata della distanza dal fiume:

```r
# coefficiente di correlazione tra i valori di zinco e la distanza dal fiume:
cor_transf = cor(cbind(log(meuse$zinc),sqrt(meuse$dist)))
plot(log(zinc)~sqrt(dist), meuse)           # scatter dei punti Zn vs DIST
abline(lm(log(zinc)~sqrt(dist), meuse))     # linea di tendenza
title( paste("Correlazione ",round(cor_transf[2],3)) )
```

#### Analisi della presenza di "struttura spaziale" nei dati.<br>
Ossia costruzione del variogramma sperimentale ed osservazione (grafica o numerica) dell'andamento della varianza ($gamma, \gamma$) all'aumentare della distanza nelle coppie di punti.

```r
lzn.vgm = variogram(log(zinc)~1, meuse) # variogramma sperimentale, log()
lzn.vgm # stampa i valori di distanza / varianza medi delle coppie di punti
```

<b>Si esegue il fitting del variogramma sperimentale (lzn.vgm) con un modello teorico ammissibile.</b><br>
Abbiamo visto nella precedente lezione diversi modelli ('Sph', 'Exp', ...).<br>
I parametri {sill,nugget,range} del variogramma sperimentale devono essere passati alla funzione fit.variogram() che crea il variogramma di fitting.<br>
In R+GSTAT, il passaggio dei parametri è mediato dalla funzione vgm(sill,model,range,nugget).<br>
In definitiva, la funzione di fitting del variogramma è così fatta:<br>
 - fit.variogram( exp_vgm, mod_vgm, ... )
 - ... indica altri parametri, come appresso indicato
 - fit.sills [DEFAULT = TRUE]
 - fit.ranges [DEFAULT = TRUE]
 - fit.method [DEFAULT = 7]

```r
lzn.fit = fit.variogram(lzn.vgm,        # variogramma sperimentale di cui eseguire il fitting
          model = vgm(1, "Sph", 900, 1))# modello di variogramma usato per il fitting
lzn.fit # stampa i parametri del modello
```

```r
options(repr.plot.width=5,repr.plot.height=4)
plot(lzn.vgm, lzn.fit)
```

Invece della media costante - denotata da <b>~1</b>, possiamo specificare una funzione media, ad esempio utilizzando <b>~sqrt(dist)</b> come predittore:

```r
lznr.vgm = variogram(log(zinc)~sqrt(dist), meuse)
lznr.fit = fit.variogram(lznr.vgm, model = vgm(1, "Exp", 300, 1))
lznr.fit
```

```r
plot(lznr.vgm, lznr.fit)
```

### Interpolazione spaziale mediante kriging:
- ordinary kriging [~1]
- universal kriging [~sqrt(dist)]

```r
lzn.kriged = krige(log(zinc)~1, meuse, meuse.grid, model = lzn.fit)
lznr.kriged = krige(log(zinc)~sqrt(dist), meuse, meuse.grid, model = lznr.fit)
```

```r
options(repr.plot.width=4)
#par( mfrow = c( 2, 2 ) ) # divide lo spazio dei grafici in 2 righe e 2 colonne
spplot(lzn.kriged["var1.pred"],main="Ordinary Kriging [~1]")
spplot(lznr.kriged["var1.pred"],main="Universal Kriging [~sqrt(dist)]")
```
